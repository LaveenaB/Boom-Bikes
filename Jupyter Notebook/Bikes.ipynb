# Feature Engineering

```python
Bikes_df.columns
```

```python
Bikes_df.head()
```

```python
# make one column of binary nos. provides if Holiday and Working is there or not
Bikes_df.loc[(Bikes_df.holiday + Bikes_df.workingday) < 1, 'days'] = 0
Bikes_df.loc[(Bikes_df.holiday + Bikes_df.workingday) > 0, 'days'] = 1
```

```python
Bikes_df
```

```python
# now as we have new column company for sibsp and parch, we can drop sibsp and parch
# data = data.drop(columns='sibsp')
data = Bikes_df.drop(columns=['instant','dteday','atemp','windspeed','holiday','workingday'])
```

```python
data
```

```python
data.describe()
```

# Data Modelling
Training the model

```python
data = data.reset_index()
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
# x is the variable of columns or data that we need the result from
x = data.drop(columns=['index','cnt','weathersit','temp','hum'], axis=1)
```

```python
# y is the variable of column or data that we need the result for
# like we need age data that what was the age of persons in the ship who survived here the condition is survived
y = data['cnt']
```

```python
# Splitting the dataset into the Training set and Test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# fitting multiple linear regression to training set
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
x_train.head()
```

```python
y_test.head(20)
```

```python
# Compute model slope and intercept
a = regressor.coef_
b = regressor.intercept_
print("Estimated model slope, a:" , a)
print("Estimated model intercept, b:" , b) 
```

# Predict the test set result

```python
# Predicting the test set results
y_pred = regressor.predict(x_test)
```

```python
# Testing the result with actual values 

regressor.predict([[2,0,6,6,1869,3473,0.0]])
```

```python
y_pred[:10]
```

# Evaluate the model

```python
# error in prediction
import math
rmse = math.sqrt(mean_squared_error(y_test,y_pred))
```

```python
rmse
```

```python
# measure the performance by r2 seems it is a good score
R2=r2_score(y_test,y_pred)
```

```python
R2
```
